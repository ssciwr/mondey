### Add STARTED dates from data.csv to MilestoneAnswerSessions in database

```python
import sqlite3 as lite

import pandas as pd
```

```python
df = pd.read_csv(
    "../mondey_backend/src/mondey_backend/import_data/data.csv",
    encoding="utf-16",
    sep="\t",
)
```

```python
df = (df.loc[:, ["CASE", "STARTED"]]).set_index("CASE")
```

```python
df.head()
```

```python
con = lite.connect("mondey.db")
```

```python
with con:
    cur = con.cursor()
    for child_id, row in df.iterrows():
        print(child_id, row.STARTED)
        cur.execute(
            f"UPDATE milestoneanswersession SET created_at='{row.STARTED}' WHERE child_id={child_id}"
        )
```
