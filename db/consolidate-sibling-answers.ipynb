### Consolidate sibling answers in database

```python
import sqlite3 as lite
```

```python
con = lite.connect("mondey.db")
```

```python
# get child ids
with con:
    cur = con.cursor()
    child_ids = [c[0] for c in cur.execute("select id from child")]
```

```python
# add 0 for younger/older sibling count if no existing answer
for child_id in child_ids:
    with con:
        cur = con.cursor()
        for question_id in [17, 18]:
            if not cur.execute(
                f"select * from childanswer where child_id={child_id} and question_id={question_id}"
            ).fetchone():
                sql = f"insert into childanswer values (0, NULL, {child_id}, {question_id});"
                print(sql)
                cur.execute(sql)
```
